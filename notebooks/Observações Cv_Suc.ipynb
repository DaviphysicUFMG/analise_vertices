```python
df00 = pd.read_csv('/home/davib/Área de Trabalho/Area de Trabalho 2023 Backup/Analise/Teste/Testar MC/Inicial/Kagome_06x06_BC_0000/Resultados/MC/Simula_0001/temps.dat', sep='\s+')
df90 = pd.read_csv('/home/davib/Área de Trabalho/Area de Trabalho 2023 Backup/Analise/Teste/Testar MC/Inicial/Kagome_06x06_BC_9000/Resultados/MC/Simula_0001/temps.dat', sep='\s+')

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
kwargs = {
    'marker': '.',
    'ms': 5,
    'ls': '--',
    'aa': True,
}

font_dict = {'fontsize': 10, 'verticalalignment':'top'}

ax[0].plot(df00.Temp/5, df00.SSF, label='Spin-flip Acceptance Ratio', **kwargs)
ax[0].plot(df00.Temp/5, df00.Kagome, label='Kagome-loop Acceptance Ratio', **kwargs)
ax[0].plot(df00.Temp/5, df00.Kag_try, label='Kagome-loop Generation Ratio', **kwargs)
ax[0].set_xscale('log')
ax[0].set_xlim(0.03, 100)
ax[0].set_ylim(-0.01,1.01)
ax[0].legend(fontsize=7)#framealpha=0.3, bbox_to_anchor=(0.8, 1.25))
ax[0].set_title('Kagome Lattice')
ax[0].set_xticks([0.03, 0.1, 1, 10, 100])
ax[0].set_xticklabels([r'', r'0.1', r'1', r'10', r'100'], fontdict=font_dict)

ax[1].plot(df90.Temp/5, df90.SSF, label='Spin-flip Acceptance Ratio', **kwargs)
ax[1].plot(df90.Temp/5, df90.Triangular, label='Triangular-loop Acceptance Ratio', **kwargs)
ax[1].plot(df90.Temp/5, df90.Tri_try, label='Triangular-loop Generation Ratio', **kwargs)
ax[1].set_xscale('log')
ax[1].set_xlim(0.03, 100)
ax[1].set_ylim(-0.01,1.01)
ax[1].legend(fontsize=7)#,framealpha=0.3, bbox_to_anchor=(0.8, 1.25))
ax[1].set_title('Triangular Lattice')
ax[1].set_xticks([0.03, 0.1, 1, 10, 100])
ax[1].set_xticklabels([r'', r'0.1', r'1', r'10', r'100'], fontdict=font_dict)

ax[0].set_xlabel("$T$ / $J_{nn}$", fontsize=15)
ax[1].set_xlabel("$T$ / $J_{nn}$", fontsize=15)
plt.show()
```
