```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('temps.dat', sep='\s+', skiprows=[0], header=None)
plt.figure(figsize=(6, 5))

plt.plot(df[0]/5, df[1], label='Spin-flip Acceptance Ratio', marker='^', ms=5)
plt.plot(df[0]/5, df[2], label='Kagome-loop Acceptance Ratio', marker='h', ms=5)

plt.hlines(1, 0.005, 100, color='r')
plt.xlabel("$T$ / $J_{nn}$")
plt.ylabel("acceptance ratios")
plt.xlim(0.005, 100)

plt.legend(framealpha=0.3)
plt.xscale('log')
plt.savefig('acc.png', bbox_inches=False)
```
