```python
from main import VerticeAnalysis
from plot import plot_vertice
import pandas as pd
import matplotlib.pyplot as plt
```

```python
config0 = pd.read_csv('config0.csv')

x = config0.x.to_numpy()
y = config0.y.to_numpy()
mx = config0.mx.to_numpy()
my = config0.my.to_numpy()
```

```python
model = VerticeAnalysis(x, y, mx, my)

Bx0 = model.Bx.copy()
By0 = model.By.copy()
charge0 = model.charge.copy()

kwargs = {
    'cmap': 'flatspin'
}

ax = plt.gca()
quiv = model.plot_model(ax=ax, **kwargs)
plot_vertice(model.v_x, model.v_y, charge0, ax=ax)

```

```python
spins = pd.read_parquet('Spin.parquet')
spins = spins.T
spins.head()
```

```python
model.Spin = spins[0]
model.calc_field()
model.get_charge()

fig, axes = plt.subplots(1, 2, figsize=(20, 9))
quiv = model.plot_model(ax=axes[0], **kwargs)
axes[1].streamplot(model.XX, model.YY, model.Bx-Bx0, model.By-By0, density=2, linewidth=1)
plot_vertice(model.v_x, model.v_y, model.charge-charge0, ax=axes[1])
axes[1].axis('equal')
```

```python
# Cálculo da carga média
import numpy as np

Spin = pd.read_parquet('Spin3.parquet')
Spin = Spin.T
Spin.head()
```

```python
Spin.shape[1]
```

```python
carga_media = np.zeros(Spin.shape[0])

for i in range(Spin.shape[1]):
    model.Spin = Spin[i]
    model.get_charge()
    carga_media += model.charge
    
carga_media /= Spin.shape[1]
```

```python
plt.figure(figsize=(20, 18))
plot_vertice(model.v_x, model.v_y, carga_media, label=True)
ax = plt.gca()
ax.set_facecolor('gray')
plt.axis('equal')
```

```python
tipo_carga = pd.DataFrame(data=np.around(model3.charge,1), columns=['carga'])
tipo_carga
```

```python
# -4 < carga < -2
tipo_I = tipo_carga.query('-4 < carga < -2').rename(columns={'carga': 'One_in_Two_out'})
tipo_II = tipo_carga.query('2 < carga < 4').rename(columns={'carga': 'Two_in_One_out'})
tipo_III = tipo_carga.query('carga > 8').rename(columns={'carga': 'Three_in'})
tipo_IV = tipo_carga.query('carga < -8').rename(columns={'carga': 'Three_out'})
```
