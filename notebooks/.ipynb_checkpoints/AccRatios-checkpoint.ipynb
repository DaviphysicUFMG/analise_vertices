```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('temps.dat', sep='\s+', skiprows=[0], header=None)
fig, ax = plt.subplots(1, 1, figsize=(6, 5))

ax.plot(df[0]/5, df[1], label='Spin-flip Acceptance Ratio', marker='^', ms=5, c='k')
ax.plot(df[0]/5, df[2], label='Kagome-loop Acceptance Ratio', marker='h', ms=5, c='b')
ax.plot(df[0]/5, df[3], label='Kagome-loop Generation Ratio', marker='o', ms=5, c='r')


ax.hlines(1, 0.005, 100, color='y', alpha=0.5)
ax.hlines(0, 0.005, 100, color='y', alpha=0.5)
ax.set_xlabel("$T$ / $J_{nn}$", fontsize=15)
ax.set_ylabel("acceptance ratios", fontsize=15)
ax.set_xlim(0.005, 100)

# ax.legend(framealpha=0.3, bbox_to_anchor =(0.8, 1.27))
ax.legend(title="$J_1 = 0$",title_fontsize=20,loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_xscale('log')
# ax.set_title('Kagome sem Troca')

font_dict = {'fontsize': 15, 'verticalalignment':'top'}

# Set ticklabels

ax.set_xticks([0.01, 0.1, 1, 10, 100])
ax.set_xticklabels([r'0.01', r'0.1', r'1', r'10', r'100'], fontdict=font_dict)

ax.set_yticks([0.0, 0.2, 0.4, 0.6, 0.8, 1])
ax.set_yticklabels([r'0', r'0.2', r'0.4', r'0.6', r'0.8', r'1'], fontdict=font_dict)

# plt.savefig('acc.png', bbox_inches=False)
plt.show()
```
